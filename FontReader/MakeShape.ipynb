```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_original = cv2.imread('coemoji.jpg')
image = image_original.copy()
image = cv2.cvtColor(image, cv2.cv.CV_RGB2GRAY)
image = cv2.GaussianBlur(image,(11,11),0)
th, image = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.figure(figsize = (20,20))
plt.imshow(image, cmap='Greys_r')
```

```python
contours, hierarchy = cv2.findContours(image, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

```python
image_contours = image_original.copy()
cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 3)
```

```python
plt.figure(figsize = (20,20))
plt.imshow(image_contours)
```
